```python
data['Age_Group'] = pd.cut(data['AGE'], 
                           bins=[0, 20, 25, float('inf')], 
                           labels=['20 leh hnuailam', '21 - 25', '25 chunglam'])

sns.countplot(x='Age_Group', data=data, palette = "pastel")

# Customize the plot
plt.title('KUM')
plt.xlabel('KUM')
plt.ylabel('Count')
plt.savefig(r"D:\mizoinsights\kum.jpg", bbox_inches='tight')
```

```python
from statistics import mean
```

```python
meanAge = mean(data["AGE"])
meanAge
```

```python
from statistics import variance as var

vara = var(data["AGE"])
```

```python
import math

std = math.sqrt(vara)
std
```

```python
sns.histplot(data["AGE"], bins=10, kde=False, color='PINK', edgecolor='black')
plt.axvline(meanAge, color='red', linestyle='dashed', linewidth=1)
plt.text(meanAge + 1, 1, f'Mean: {meanAge:.2f}', color='red')

plt.axvline(meanAge + std, color='green', linestyle='dashed', linewidth=1)
plt.axvline(meanAge - std, color='green', linestyle='dashed', linewidth=1)
plt.text(meanAge + std + 1, 1, f'SD: {std:.2f}', color='green')
plt.text(meanAge - std - 3, 1, f'SD: {std:.2f}', color='green')

# Labels and title
plt.title('KUM IN SEMDARH DAN')
plt.xlabel('KUM')
plt.ylabel('Frequency')
plt.savefig(r"D:\mizoinsights\kuminsemdarh.jpg", bbox_inches='tight')
# Show plot
plt.show()
```

```python
# Clean the "MARKET VALUE" column
data["MARKET VALUE"] = data["MARKET VALUE"].replace({'\$': '', ',': ''}, regex=True).astype(float)
```

```python
meanVal = data["MARKET VALUE"].mean()
meanVal
```

```python
medianVal = data["MARKET VALUE"].median()
medianVal
```

```python
rangeVal = data["MARKET VALUE"].dropna().max() - data["MARKET VALUE"].dropna().min()
rangeVal
```

```python
varianceVal = data["MARKET VALUE"].var()
varianceVal
```

```python
stdDevVal = data["MARKET VALUE"].std()
stdDevVal
```

```python
print(f'Mean Market Value: {meanVal}')
print(f'Median Market Value: {medianVal}')
print(f'Range of Market Values: {rangeVal}')
print(f'Variance of Market Values: {varianceVal}')
print(f'Standard Deviation of Market Values: {stdDevVal}')
```

```python
# Set the style for more attractive plots
sns.set_style("whitegrid")
plt.figure(figsize=(16, 6))

# Histogram with KDE
plt.subplot(1, 2, 1)
sns.histplot(data=data, x="MARKET VALUE", color="skyblue", edgecolor="darkblue")
plt.title('Market Values Analysis', fontsize=16)
plt.xlabel('Market Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.savefig(r"D:\mizoinsights\market1.jpg", bbox_inches='tight')
plt.show()

# Box plot
sns.boxplot(x=data["MARKET VALUE"], color="lightgreen")
sns.stripplot(x=data["MARKET VALUE"], color="darkgreen", alpha=0.5, size=4)
plt.title('Market Values Analysis', fontsize=16)
plt.xlabel('Market Value', fontsize=12)
plt.savefig(r"D:\mizoinsights\market2.jpg", bbox_inches='tight')
plt.show()

# Adjust layout and display

# Display summary statistics
summary_stats = data["MARKET VALUE"].describe()
print(summary_stats)
```
